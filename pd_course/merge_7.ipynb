```python
import pandas as pd 
```

```python
week1 = pd.read_csv("pandas\pandas\Restaurant - Week 1 Sales.csv")
```

```python
week2 = pd.read_csv("pandas\pandas\Restaurant - Week 2 Sales.csv")
```

```python
customers = pd.read_csv(r"pandas\pandas\Restaurant - Customers.csv")
```

```python
foods = pd.read_csv(r"pandas\pandas\Restaurant - Foods.csv")
```

```python
pd.concat(objs=[week1, week2], ignore_index=True)
```

```python
pd.concat(objs=[week1, week2], keys=["week1", "week2"]).loc[("week1", 33)]
```

```python
week1.head()
week1.merge(week2, how="inner", on="Customer ID", suffixes=("_1", "_2")).head(4)
```

```python
week1.head(3)
week1.merge(week2, how="inner", on=["Customer ID", "Food ID"])
```

```python
merged = week1.merge(week2, how="outer", on="Customer ID", suffixes=[" - 1", " - 2"], indicator=True)
```

```python
merged["_merge"].value_counts()
```

```python
mask = merged["_merge"].isin(["left_only", "right_only"])
merged[mask]
```
