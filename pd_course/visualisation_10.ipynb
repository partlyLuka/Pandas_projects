```python
def f(x):
    if x <= 10:
        return "Poor"
    elif x <= 50:
        return "Satisfactory" 
    else:
        return "Supreme"
```

```python
s = df["Close"].apply(f)
```

```python
s.value_counts().plot.bar()
```

```python
s = s.value_counts()
```

```python
s.plot(kind="bar")
```

```python
s.plot.barh()
```

```python
avg = df["Close"].mean()
avg
```

```python
def g(x):
    if x >= avg:
        return "Above average"
    return "Below average"
```

```python
s = df["Close"].apply(g)
```

```python
s.value_counts().plot.bar()
```

```python
plt.style.use("dark_background")
```

```python
s.value_counts().plot(kind="pie", legend=True)
```
