```python
fram = pd.DataFrame(columns=df.columns)
fram
```

```python
sectors.groups
for sec in sectors.groups:
    for i in sectors.groups[sec]:
        rev = df.loc[i, "Revenue"]
        if rev == sectors["Revenue"].max()[sec]:
            fram.loc[i] = df.loc[i]
                 
```

```python
i = 0
for sec, data in sectors:
    row = data.nlargest(1, "Revenue")
    #print(row.values)
    fram.loc[i] = row.values[0]
    i += 1
fram  
```

```python
locations = df.groupby("Location")
```
